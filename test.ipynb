## 定义模型

```python
from torch import nn 
class FC(nn.Module):
    def __init__(self, input_dim, output_dim):
        super(FC, self).__init__()
        self.fc = nn.Sequential(
            nn.Linear(input_dim, 100),
            nn.ReLU(),
            nn.Linear(100, 50),
            nn.ReLU(),
            nn.Linear(50, output_dim)
        )
    def forward(self, x):
        return self.fc(x)
    
```

```python
import torch
import torch.nn as nn
import torch.optim as optim

# Create an instance of your model
model = FC(geno_array.shape[1], 1)

# Define your loss function and optimizer
criterion = nn.MSELoss()
optimizer = optim.SGD(model.parameters(), lr=1e-3)
```

```python
from torch.utils.data import TensorDataset, DataLoader

dataloader  = DataLoader(TensorDataset(geno_array, pheno), batch_size=64, shuffle=True)
dataloader 
```

```python
%%time 
model.to('cuda:0')
metrics = []
step= 0 
for epoch in range(1, 30):
    # epoch start 
    for x, y in dataloader:
        # print(x.shape, y.shape)
        y_pred = model(x.to('cuda:0'))
        loss = criterion(y_pred, y.to("cuda:0"))
        # print(loss)
        # print(model.fc[0].weight.grad.mean())

        optimizer.zero_grad() # 清楚之前的梯度
        loss.backward()
        optimizer.step()
        # print(model.fc[0].weight.grad.mean())
        if step % 50 == 0:
            metrics.append({"epoch":epoch, "step":step, "loss":loss.item()})

        step += 1
        
```

```python
import seaborn as sns 
res = pd.DataFrame(metrics)
sns.lineplot(data=res, x='step', y='loss')
```

```python
y_pred = model(geno_array.to('cuda:0'))
y_pred 
```

```python
import matplotlib.pyplot as plt
fig, ax = plt.subplots(figsize=(10, 10))
sns.scatterplot(x=pheno.squeeze().numpy(), y=y_pred.squeeze().detach().cpu().numpy(), ax = ax )

ax.set_xlabel("real ANGPTL3 values")
ax.set_ylabel("predicted ANGPTL3 values")
```

```python
import statsmodels.api as sm

ols = sm.OLS(pheno.squeeze().numpy(), sm.add_constant(y_pred.squeeze().detach().cpu().numpy()))
ols_res = ols.fit()
```

```python
ols_pred = ols_res.predict(sm.add_constant(y_pred.squeeze().detach().cpu().numpy()))
ols_pred
```
